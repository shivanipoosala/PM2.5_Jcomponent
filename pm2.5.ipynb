```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv("D:\python\pm.csv")
df
```

```python
dfe=df.replace(to_replace="Na",value=0)
ds=dfe.replace(to_replace=np.NaN,value=0)
ds
```

```python
reg =linear_model.LinearRegression()
reg.fit(ds[['DEWP','TEMP','PRES','Iws']],ds.pm)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
a=float(input("Enter the DEW POINT="))
b=float(input("Enter the TEMPERATURE="))
c=float(input("Enter the PRESSURE="))
d=float(input("Enter the WIND SPEED="))
```

```python
pm=reg.predict([[a,b,c,d]])
```

```python
print('pm2.5=',pm)
```
